```python
import pickle 
T = 5
fname = "playground-1606807028.pickle"
train = []
with open(fname, "rb") as f:
    us, xs = pickle.load(f)
    for t in range(T, len(xs[0])-T, T):
        point = {
            'x0': [xseq[t-1] for xseq in xs],
            'u': [useq[t:t+T] for useq in us]
        }
        train.append(point)

        
```
